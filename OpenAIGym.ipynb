```python
#Macros
LEFT = 0 
DOWN = 1
RIGHT = 2
UP = 3

#Key mapping
arrow_keys = {
    75:LEFT,
    77:RIGHT,
    72:UP,
    80:DOWN
}
```

```python
register(
    id ='FrozenLake-v4',
    entry_point='gym.envs.toy_text:FrozenLakeEnv',
    kwargs={'map_name' : '4x4' , 'is_slippery': False}
)
```

```python
#gym에서 제공하는 어떤 환경을 하나 만듬 -> env
env = gym.make('FrozenLake-v0') 
observation = env.reset() #초기화

for _ in range(1000) : 
    env.render() #환경의 화면을 출력
    action = env.action_space.sample()  #랜덤으로 엑션을 만듬 
    obsevation , reward , done , info = env.step(action) #done = 끝이 났는지 , info = 추가정보
```

```python
env = gym.make('FrozenLake-v4')
env.render()

while True :
    if ord(msvcrt.getch()) != 224:          # 특수키 입력은 224
        print("Game aborted!, not arrow")
        break

    key = ord(msvcrt.getch())
    
    if key not in arrow_keys.keys() : 
        print('Game aborted!')
        break
        
    action = arrow_keys[key]
    state , reward, done, info = env.step(action)
    env.render()
    
    print('state : ' , state , 'Action : ' ,action , 'Reward : ' , reward , 'Info : ' , info )
    
    if done :
        print('Finished with reward' , reward)
        break
```
